```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
TAG_NAME = "classification_loss"
tf_event_file = '/Users/mhendriksen/Desktop/repositories/SCAN/runs/coco_scan/log/events.out.tfevents.1639066754.LMAC-X61SPJGH5'

value_list = []
for e in tf.compat.v1.train.summary_iterator(tf_event_file):
    for v in e.summary.value:
        if v.tag == TAG_NAME:
            value = tf.make_ndarray(v.tensor)
            value_list.append(value)

plt.plot(value_list)
plt.grid()
plt.show()
```
